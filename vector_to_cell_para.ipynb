```python
import numpy as np
```

```python
#input cell parameter (cell vetors) as in QE file, output unit cell parameters
a1 = np.array([   25.9850000000 ,  0  , 0])
a2 = np.array([   -12.9925000000 , 22.5036701173  , 0])
a3 = np.array([  0 ,0 , 3])
```

```python
alpha = np.rad2deg(np.arccos(np.dot(a2,a3)/(np.linalg.norm(a2)*np.linalg.norm(a3))))
beta = np.rad2deg(np.arccos(np.dot(a1,a3)/(np.linalg.norm(a1)*np.linalg.norm(a3))))
gamma = np.rad2deg(np.arccos(np.dot(a1,a2)/(np.linalg.norm(a1)*np.linalg.norm(a2))))
a = np.linalg.norm(a1)
b = np.linalg.norm(a2)
c = np.linalg.norm(a3)
print('alpha = ', alpha)
print('beta = ', beta)
print('gamma = ', gamma)
print('a = ', a)
print('b = ', b)
print('c = ', c)
```

```python
#This part of the code gives d spacing, based on hkl and their lattice type
dhkl = np.array([1,0,1])
```

```python
d_cubic = np.sqrt((np.square(a))/(np.square(dhkl[0]) + np.square(dhkl[1]) + np.square(dhkl[2])))
print('d_spacing ', dhkl[0], dhkl[1], dhkl[2])
print ('\u212B = ', d_cubic)
```

```python
d_hexgonal = 1/ (np.sqrt( ((4/3) * ((np.square(dhkl[0]) + np.square(dhkl[1]) + (dhkl[1] * dhkl[0]) ) / (np.square(a)) )+ (np.square(2) / np.square (c)))))
print('d_spacing ', dhkl[0], dhkl[1], dhkl[2])
print ( d_hexgonal, '\u212B ')
```

```python
d_tetragonal = 1 / (np.sqrt((np.square(dhkl[0]) + np.square(dhkl[1])) / (np.square(a)) + (np.square(2) / np.square(c))))
print('d_spacing ', dhkl[0], dhkl[1], dhkl[2])
print (d_tetragonal , '\u212B')
```
