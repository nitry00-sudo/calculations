```python
import numpy as np
```

```python
data = np.loadtxt('FULL/PATH/TO/DATA/filename.band.pp.dat.gnu') #path of .gnu file
```

```python
split_data = np.hsplit(data,2)
x_coord = np.unique(split_data[0])
y_coord = split_data[1]
print(x_coord)
```

```python
rows = np.shape(x_coord)[0]
columns = np.shape(y_coord)[0]//rows
print(rows, columns)
```

```python
Y = y_coord.reshape(rows, columns)
```

```python
final_data = np.insert(Y,0,x_coord,axis=1)
print(final_data)
```

```python
np.savetxt('FULL/PATH/TO/DATA/filename.csv',final_data,delimiter=',') #path to final data
```
