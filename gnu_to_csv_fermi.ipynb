```python
import numpy as np
```

```python
data = np.loadtxt('/Users/charlie/Downloads/BTTL1.band.pp.dat.gnu')  #path of .gnu file
split_data = np.hsplit(data, 2)
x_coord = np.unique(split_data[0])
y_coord = split_data[1]
rows = np.shape(x_coord)[0]
columns = np.shape(y_coord)[0] // rows
Y = y_coord.reshape(rows, columns)
final_data = np.insert(Y, 0, x_coord, axis=1)
```

```python
Fermi = 3   #insert your fermi level
```

```python
farry = np.zeros((rows,columns))
```

```python
farry.fill(Fermi)
```

```python
print(farry)
```

```python
np.insert(farry, 0, 0, axis=1)
```

```python
fermi_data = np.insert(farry, 0, 0, axis=1) + final_data
```

```python
np.savetxt('/Users/charlie/Downloads/BTTL1.band.csv',fermi_data,delimiter=',') #path to final data
```
